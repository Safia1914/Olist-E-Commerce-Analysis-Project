**Merge DataFrames to Build a Master Dataset**

This step combines all tables into one master dataset for analysis.

Merge 1: Orders + Customers

```python
order_customer=pd.merge(orders,customers,on='customer_id')
order_customer.head()
```

Merge 2: Add Payments

```python
order_payment = pd.merge(order_customer, payments, on='order_id')
order_payment.head()
```

Merge 3: Add Order Items

```python
order_full = pd.merge(order_payment, order_items, on='order_id')
order_full.head()
```

Merge 4: Add Product Details


```python
order_full = pd.merge(order_full, products, on='product_id')
order_full.head()
```

Merge 5: Translate Product Categories

```python
order_full = pd.merge(order_full, category_translation, on='product_category_name', how='left')
#Translates Portuguese category names into English using the category_translation table.how='left' keeps all rows from order_full, even if some categories aren’t translated.
order_full.head()
```

# B. RFM Segmentation (Recency, Frequency, Monetary)

```python
import datetime as dt

snapshot_date = order_full['order_purchase_timestamp'].max() + dt.timedelta(days=1)
rfm_df = order_full.groupby('customer_unique_id').agg({
    'order_purchase_timestamp': lambda x: (snapshot_date - x.max()).days,
    'order_id': 'count',
    'payment_value': 'sum'
}).reset_index()

rfm_df.columns = ['customer_id', 'Recency', 'Frequency', 'Monetary']
```

# C. Segment with Quantiles

```python
rfm_df['R_Score'] = pd.qcut(rfm_df['Recency'], 4, labels=[4, 3, 2, 1])
rfm_df['F_Score'] = pd.qcut(rfm_df['Frequency'].rank(method='first'), 4, labels=[1, 2, 3, 4])
rfm_df['M_Score'] = pd.qcut(rfm_df['Monetary'], 4, labels=[1, 2, 3, 4])

rfm_df['RFM_Segment'] = rfm_df['R_Score'].astype(str) + rfm_df['F_Score'].astype(str) + rfm_df['M_Score'].astype(str)
rfm_df['RFM_Score'] = rfm_df[['R_Score', 'F_Score', 'M_Score']].sum(axis=1)
```
